# Model Building

```python
X = df.drop("ProdTaken",axis = 1)
y = df["ProdTaken"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.30,random_state=99)
```

```python
num_features = X.select_dtypes(exclude='object').columns
cat_features = X.select_dtypes(include='object').columns
```

```python
num_transform = StandardScaler()
ohe = OneHotEncoder()
CT = ColumnTransformer(
    [
        ("OneHotEncoder",ohe,cat_features),
        ("StandardScaler",num_transform,num_features)
    ]
)
```

```python
CT
```

```python
X_train = CT.fit_transform(X_train)
```

```python
X_test = CT.transform(X_test)
```

```python
pd.DataFrame(X_train)
```

```python
models={
    "Logisitic Regression":LogisticRegression(),
    "Decision Tree":DecisionTreeClassifier(),
    "Random Forest":RandomForestClassifier(),
    "KNN":KNeighborsClassifier(),
    "SVC":SVC(class_weight='balanced'),
}
for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train, y_train) # Train model

    # Make predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Training set performance
    model_train_accuracy = accuracy_score(y_train, y_train_pred) # Calculate Accuracy
    model_train_f1 = f1_score(y_train, y_train_pred, average='weighted') # Calculate F1-score
    model_train_precision = precision_score(y_train, y_train_pred) # Calculate Precision
    model_train_recall = recall_score(y_train, y_train_pred) # Calculate Recall
    model_train_rocauc_score = roc_auc_score(y_train, y_train_pred)


    # Test set performance
    model_test_accuracy = accuracy_score(y_test, y_test_pred) # Calculate Accuracy
    model_test_f1 = f1_score(y_test, y_test_pred, average='weighted') # Calculate F1-score
    model_test_precision = precision_score(y_test, y_test_pred) # Calculate Precision
    model_test_recall = recall_score(y_test, y_test_pred) # Calculate Recall
    model_test_rocauc_score = roc_auc_score(y_test, y_test_pred) #Calculate Roc


    print(list(models.keys())[i])
    
    print('Model performance for Training set')
    print("- Accuracy: {:.4f}".format(model_train_accuracy))
    print('- F1 score: {:.4f}'.format(model_train_f1))
    
    print('- Precision: {:.4f}'.format(model_train_precision))
    print('- Recall: {:.4f}'.format(model_train_recall))
    print('- Roc Auc Score: {:.4f}'.format(model_train_rocauc_score))

    
    
    print('----------------------------------')
    
    print('Model performance for Test set')
    print('- Accuracy: {:.4f}'.format(model_test_accuracy))
    print('- F1 score: {:.4f}'.format(model_test_f1))
    print('- Precision: {:.4f}'.format(model_test_precision))
    print('- Recall: {:.4f}'.format(model_test_recall))
    print('- Roc Auc Score: {:.4f}'.format(model_test_rocauc_score))

    
    print('='*35)
    print('\n')
```

# HyperParameter Tuning

```python
rf_params = {
    "max_depth":[1,10,15,30,50,None],
    "max_features":[3,6,9,'auto'],
    "min_samples_split":[5,10,13,15,20],
    "n_estimators":[100,200,300,400,500]
}

knn_params = {
    "n_neighbors":[5,10,15,20,30],
    "weights":["uniform","distance"],
    "metric":["euclidean","manhattan"]
}
```

```python
rf_params
```

```python
knn_params
```

```python
randomcv_models = [
    ("RF",RandomForestClassifier(),rf_params,75,5),
    ("KNN",KNeighborsClassifier(),knn_params,15,5)
]
randomcv_models
```

```python
model_params = {}
for name,model,params,n_iter,cv in randomcv_models:
    random = RandomizedSearchCV(
        estimator=model,
        param_distributions=params,
        n_jobs=-1,
        cv=cv,
        n_iter=n_iter,
        verbose=2
    )
    random.fit(X_train,y_train)
    model_params[name]=random.best_params_
for model_name in model_params:
    print(f"---------------- Best Params for {model_name} -------------------")
    print(model_params[model_name])
```

```python
models={
    
    "Random Forest":RandomForestClassifier(n_estimators=300,min_samples_split=5,
                                          max_features=9,max_depth=None),
    "KNN":KNeighborsClassifier(weights='distance',n_neighbors=5,metric='euclidean')
}
for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train, y_train) # Train model

    # Make predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Training set performance
    model_train_accuracy = accuracy_score(y_train, y_train_pred) # Calculate Accuracy
    model_train_f1 = f1_score(y_train, y_train_pred, average='weighted') # Calculate F1-score
    model_train_precision = precision_score(y_train, y_train_pred) # Calculate Precision
    model_train_recall = recall_score(y_train, y_train_pred) # Calculate Recall
    model_train_rocauc_score = roc_auc_score(y_train, y_train_pred)


    # Test set performance
    model_test_accuracy = accuracy_score(y_test, y_test_pred) # Calculate Accuracy
    model_test_f1 = f1_score(y_test, y_test_pred, average='weighted') # Calculate F1-score
    model_test_precision = precision_score(y_test, y_test_pred) # Calculate Precision
    model_test_recall = recall_score(y_test, y_test_pred) # Calculate Recall
    model_test_rocauc_score = roc_auc_score(y_test, y_test_pred) #Calculate Roc


    print(list(models.keys())[i])
    
    print('Model performance for Training set')
    print("- Accuracy: {:.4f}".format(model_train_accuracy))
    print('- F1 score: {:.4f}'.format(model_train_f1))
    
    print('- Precision: {:.4f}'.format(model_train_precision))
    print('- Recall: {:.4f}'.format(model_train_recall))
    print('- Roc Auc Score: {:.4f}'.format(model_train_rocauc_score))

    
    
    print('----------------------------------')
    
    print('Model performance for Test set')
    print('- Accuracy: {:.4f}'.format(model_test_accuracy))
    print('- F1 score: {:.4f}'.format(model_test_f1))
    print('- Precision: {:.4f}'.format(model_test_precision))
    print('- Recall: {:.4f}'.format(model_test_recall))
    print('- Roc Auc Score: {:.4f}'.format(model_test_rocauc_score))

    
    print('='*35)
    print('\n')
```

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    filename = name.replace(" ", "_").lower() + ".pkl"  # e.g., 'random_forest.pkl'
    joblib.dump(model, filename)
    print(f"{name} saved as {filename}")
```
